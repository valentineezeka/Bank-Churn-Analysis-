# 📊 What is EDA (Exploratory Data Analysis)?

Exploratory Data Analysis (EDA) is the process of examining and summarizing a dataset to understand its main characteristics **before drawing any conclusions**.

Think of it as **getting to know your data** — you look at the shape, patterns, trends, and any oddities before jumping into deeper analysis.

## 🔍 What We Do in EDA:
- Look at the **structure** of the data (rows, columns, data types)
- Check for **missing values** and **outliers**
- Understand how variables are **distributed**
- Find **relationships** between columns
- Use **statistics** (mean, median, min, max) and **visualizations** (bar charts, box plots, pie charts, scatter plots)

## 🎯 Goal of EDA:
> *EDA does not give final answers — it guides our analysis, spots problems early, and helps us tell a story with the data.*

## 🌍 2. Group Analysis — Churn by Geography
*Which country loses the most customers?*

```python
print("\n== Churn by Geography ==")
print(df.groupby("Geography")["Exited"].agg(["mean","sum","count"])
        .rename(columns={"mean":"Churn Rate","sum":"Churned","count":"Total"})
        .assign(**{"Churn Rate": lambda x: (x["Churn Rate"]*100).round(1)})
        .to_string())

print("\n== Churn by Age Group ==")
print(df.groupby("AgeGroup", observed=True)["Exited"].mean().mul(100).round(1))
```

## 📈 3. Data Visualizations
*Minimum 4 charts to explore patterns, distributions, and relationships.*

### 📊 Chart 1 — Churned Customers by Geography (Bar Chart)
*Comparing the raw number of churned customers across countries.*

```python
import matplotlib.pyplot as plt

# Create geo_analysis dataframe
geo_analysis = df.groupby("Geography")["Exited"].sum().to_frame(name="Churned Customers")

# Bar Chart
geo_analysis["Churned Customers"].plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Churned Customers by Geography")
plt.xlabel("Geography")
plt.ylabel("Number of Churned Customers")
plt.xticks(rotation=0)
plt.show()
```

### 📈 Chart 2 — Customer Churn Trend Over Time (Line Chart)
*How has churn changed year by year between 2020–2025?*

```python
# Churn trend by year
yearly_churn = df.groupby("Year")["Exited"].sum()

# Plot
plt.figure(figsize=(8,5))
plt.plot(yearly_churn.index, yearly_churn.values, marker='o')

plt.title("Customer Churn Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Churned Customers")

plt.grid(True)
plt.show()
```

### 🥧 Chart 3 — Churn Distribution (Pie Chart)
*What proportion of customers churned vs. stayed?*

```python
# Count values
churn_distribution = df["Exited"].value_counts()

# Labels
labels = ["Retained", "Churned"]

# Plot
plt.figure(figsize=(6,6))
plt.pie(
    churn_distribution,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
age_exit = pd.crosstab(df['AgeGroup'], df['Exited'], normalize='index') * 100

age_exit.plot(
    kind='bar',
    stacked=True,
    figsize=(9,5)
)

plt.title("Customer Churn Percentage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Percentage")

plt.legend(['Retained','Churned'])

plt.xticks(rotation=0)

plt.show()
```

```python
#Customers aged 46–65 have the highest churn rates, exceeding 48%,
#while younger customers below 35 show churn rates below 10%.
#This indicates age is a significant predictor of customer churn.
```

```python
age_churn = df.groupby('AgeGroup')['Exited'].mean() * 100


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))

ax = sns.barplot(
    x=age_churn.index,
    y=age_churn.values
)

# Add labels
for i, value in enumerate(age_churn.values):
    ax.text(
        i,
        value + 1,
        f'{value:.1f}%',
        ha='center',
        fontweight='bold'
    )

plt.title("Customer Churn Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Churn Rate (%)")

plt.ylim(0,60)

plt.show()
```

```python
# Customer churn shows a clear upward trend with age.
# Customers aged 46–55 have the highest churn rate (50.6%),
# followed by customers aged 56–65 (48.3%).
# Meanwhile, customers below 35 years maintain churn rates below 10%.
# This indicates that older customers are significantly more likely to leave the bank,
# making age a key factor influencing customer retention.
```

```python
import matplotlib.pyplot as plt

active_churn = df.groupby("IsActiveMember")["Exited"].mean() * 100

plt.figure(figsize=(6,5))

bars = plt.bar(["Inactive", "Active"], active_churn.values)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1,
             f"{height:.1f}%", ha='center', fontweight='bold')

plt.title("Churn Rate: Active vs Inactive Members")
plt.ylabel("Churn Rate (%)")
plt.ylim(0, max(active_churn.values) + 10)

plt.show()
```

```python
# Active members are significantly less likely to churn compared to inactive members.
# This suggests that customer engagement plays an important role in retention,
# and increasing customer activity may help reduce churn.
```
